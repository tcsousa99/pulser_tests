```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfp = pd.read_csv("pulser_proton_peaks.csv")
dfg = pd.read_csv("pulser_gamma_peaks.csv")
print(dfg.to_string())
```

### Proximo passo é fazer a razão entre os valores destas duas tabelas e fazer plots. Para a mesma tensão, cada cristal deve ter uma razão.

```python
R1 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(0, len(dfg), 16)])
R2 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(1, len(dfg), 16)])
R3 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(2, len(dfg), 16)])
R4 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(3, len(dfg), 16)])
R5 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(4, len(dfg), 16)])
R6 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(5, len(dfg), 16)])
R7 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(6, len(dfg), 16)])
R8 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(7, len(dfg), 16)])
R9 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(8, len(dfg), 16)])
R10 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(9, len(dfg), 16)])
R11 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(10, len(dfg), 16)])
R12 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(11, len(dfg), 16)])
R13 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(12, len(dfg), 16)])
R14 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(13, len(dfg), 16)])
R15 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(14, len(dfg), 16)])
R16 = np.array([(dfg.iloc[i].Mean/dfp.iloc[i].Mean) for i in range(15, len(dfg), 16)])
```

```python
voltages = [0.5,0.6,0.7,0.8,0.9,1.0]
```

```python
plt.figure(figsize=(10,8),dpi=150)



plt.scatter(voltages,R1, label='Channel 1', alpha= 0.7)
plt.scatter(voltages,R2, label='Channel 2', alpha= 0.7)
plt.scatter(voltages,R3, label='Channel 3', alpha= 0.7)
plt.scatter(voltages,R4, label='Channel 4', alpha= 0.7)
plt.scatter(voltages,R5, label='Channel 5', alpha= 0.7)
""" plt.scatter(voltages,R7, label='Channel 7', alpha= 0.7)
plt.scatter(voltages,R8, label='Channel 8', alpha= 0.7)
plt.scatter(voltages,R9, label='Channel 9', alpha= 0.7)
plt.scatter(voltages,R10, label='Channel 10',alpha=0.7)
plt.scatter(voltages,R11, label='Channel 11',alpha=0.7)
plt.scatter(voltages,R12, label='Channel 12',alpha=0.7)
plt.scatter(voltages,R13, label='Channel 13',alpha=0.7)
plt.scatter(voltages,R14, label='Channel 14',alpha=0.7)
plt.scatter(voltages,R15, label='Channel 15',alpha=0.7)
plt.scatter(voltages,R16, label='Channel 16',alpha=0.7)
 """

plt.xlabel('Pulser Voltage (V)')
plt.ylabel('Ratio Gamma/Proton Centroids')
plt.legend(loc = 'upper right', fontsize = 'small')
plt.show()
```

```python
G1 = np.array([(dfg.iloc[i].Mean) for i in range(2, len(dfg), 16)])
```

### Gamma range response saturates after 0.7V

```python
plt.figure(figsize=(6,4),dpi=150)
plt.xlabel('Pulser Voltage (V)')
plt.ylabel('Gamma Centroid Position')
plt.scatter(voltages, G1)
plt.show()
```
